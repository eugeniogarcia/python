```javascript
//Funciones auxiliares para manejar fechas.
//Convertir a fecha y ordenar por fecha
convert=function(d) {
        // Converts the date in d to a date-object. The input can be:
        //   a date object: returned without modification
        //  an array      : Interpreted as [year,month,day]. NOTE: month is 0-11.
        //   a number     : Interpreted as number of milliseconds
        //                  since 1 Jan 1970 (a timestamp) 
        //   a string     : Any format supported by the javascript engine, like
        //                  "YYYY/MM/DD", "MM/DD/YYYY", "Jan 31 2009" etc.
        //  an object     : Interpreted as an object with year, month and date
        //                  attributes.  **NOTE** month is 0-11.
        return (
            d.constructor === Date ? d :
            d.constructor === Array ? new Date(d[0],d[1],d[2]) :
            d.constructor === Number ? new Date(d) :
            d.constructor === String ? new Date(d) :
            typeof d === "object" ? new Date(d.year,d.month,d.date) :
            NaN
        )
}

compara=function(a,b) {
        // Compare two dates (could be of any type supported by the convert
        // function above) and returns:
        //  -1 : if a < b
        //   0 : if a = b
        //   1 : if a > b
        // NaN : if a or b is an illegal date
        // NOTE: The code inside isFinite does an assignment (=).
        return (
            isFinite(a=convert(a.endingDate).valueOf()) &&
            isFinite(b=convert(b.endingDate).valueOf()) ?
            ((a<b)-(a>b)) :
            NaN
        );
    }
```

```javascript
//Ordena por fecha. Los consumos quedaran ordenados de mas reciente a menos reciente
//Sino queremos ordenar los bucles que vienen a continuación en lugar de ir desde 0 en adelante 
//deberían ir desde el final hacia atrás
sips_ordenado=sips.sipsReadings.sort(compara);
```

```javascript
//Procesa las lecturas
//Energía consumida en cada dia
var energia_dia=[];
//Variable intermedia. Numero de días que hemos procesado
var dias=0;
const dias_ano=365;

//Numero de lecturas que necesitaremos para tener al menos 365 días de consumos
var periodos=0;

for (i=0;i<sips_ordenado.length;i++){
    //En javascript la diferencia entre ts viene en milisegundos, por eso dividimos por 1000x24*60*60
    //Dias incluidos en la lectura
    dias_lectura=(convert(sips_ordenado[i].endingDate)-convert(sips_ordenado[i].startDate))/24/60/60/1000;
    consumo_medio_diario=(sips_ordenado[i].consumptionP1+sips_ordenado[i].consumptionP2+
        sips_ordenado[i].consumptionP3+sips_ordenado[i].consumptionP4+
        sips_ordenado[i].consumptionP5+sips_ordenado[i].consumptionP6)/dias_lectura;
    for(j=0;j<dias_lectura;j++){
        energia_dia[dias]=consumo_medio_diario;
        dias++;
        if(dias>=dias_ano)break;
    }
    
    periodos++
    if(dias>=dias_ano) break;
}
```
